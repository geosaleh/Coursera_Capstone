```python
import pandas as pd
import numpy as np
import geocoder # import geocoder
```

load CSV file from first part

```python
df = pd.read_csv('df_final.csv')
df.head()
```

Use the Geocoder package to update dataframe with Lat and Long:

```python
def get_lat_long(postcode):
    # initialize your variable to None
    lat_lng_coords = None

    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(postcode))
        lat_lng_coords = g.latlng

    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]
    return latitude, longitude
```

```python
df['Latitude'], df['Longitude'] = zip(*df['Postcode'].apply(get_lat_long))
```
